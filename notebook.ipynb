# jupyter notebook 的基本使用技巧

以下所有操作都需要在命令模式下完成(按ESC或者单击编辑框外的空白处)

* 向上增加空白cell:  <span style="color:blue">A</span>    
* 向下增加空白cell:  <span style="color:blue">B</span>    
* 运行cell，并将光标移动到下一个cell:  <span style="color:blue">shift+enter</span>   
* 运行本cell:  <span style="color:blue">ctrl+enter</span>   
* 删除本cell:  <span style="color:blue">DD</span>   
* 剪切本cell:  <span style="color:blue">X</span>   
* 贴本cell:  <span style="color:blue">V</span>   

* 合并两个cell:  <span style="color:blue">A</span>   
* 将cell转换成代码状态:  <span style="color:blue">Y</span>   
* 将cell转换成markdown状态:  <span style="color:blue">M</span>   
* 将选中的cell合并:  <span style="color:blue">shift+M</span>   
* 打开/关闭行号:  <span style="color:blue">L</span>   
* 恢复删除的最后一个cell:  <span style="color:blue">Z</span>   
* 代码左移/右移:  <span style="color:blue">ctrl+[ / ]</span>
